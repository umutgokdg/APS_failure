```python
from preprocessing import LDA
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=2, weights=None, flip_y=0.01, class_sep=1.0, hypercube=True, shift=0.0, scale=1.0, shuffle=True, random_state=None)
```

```python
lda = LDA(1)
```

```python
lda.fit(X, y)
```

```python
X_transformed = lda.transform(X)
```

```python
import matplotlib.pyplot as plt
```

```python
#plot the transformed data
plt.scatter(X_transformed[:,0], [0]*len(X_transformed), c=y)
plt.show()
```

```python
#import linear discriminant analysis from sklearn

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda_sklearn = LinearDiscriminantAnalysis(n_components=1)

lda_sklearn.fit(X, y)

X_transformed_sklearn = lda_sklearn.transform(X)

#create plot

plt.scatter(X_transformed_sklearn[:,0], [0]*len(X_transformed_sklearn), c=y)

plt.show()
```

```python
#plot the original data
plt.scatter(X[:,0], X[:,1], c=y)
```
